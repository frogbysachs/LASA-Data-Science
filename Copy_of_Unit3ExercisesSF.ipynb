## The First Four Moments: Mean, Variance, Skew, and Kurtosis

When we talk about a probablity dist, we often care about at four things:
- The expected value, or where a distribution is centered (mean)
- The amount of variation from that centered value (variation)
- Whether or not our distribution favors, leans towards, or skews towards certain values. (skew)
- How often to expect unusually high or low values. (kurtosis)

**Task10**:

Describe what process each of the previous discrete distributions models, and their support.

Hint: Make sure to read any of the preliZ documentation, the wikipedia articles, or wolfram alpha.


*Bernoulli distribution models the probability of successes vs failures and only supports 0 or 1*

*Binomial distribution models the independent number of successes and failures in a bunch of experiments and supports all natural numbers*

*Poisson distribution models the number of events occurring at distinct points in time and supports all natural numbers*

*Geometric distribution models the probability that the first success in a number of Bernoulli trials is on the xth trial and supports all counting numbers*

*Hypergeometric distribution models the number of successes picking k good objects out of N total objects and supports x in [max(0,n-N+k),min(n,k)].*

*Categorical distribution models any discrete set of data and supports the natural numbers up to (p-1).*
